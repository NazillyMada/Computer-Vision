# Thresholding

## Global Threshold

```
average1 = cv.blur(equalized1,(3,3))
average2 = cv.blur(equalized2,(3,3))
average3 = cv.blur(equalized3,(3,3))
average4 = cv.blur(equalized4,(3,3))
average5 = cv.blur(equalized5,(3,3))
```

```
threshold_value1, threshold_result1 = cv.threshold(average1, 127, 255, cv.THRESH_BINARY)
threshold_value2, threshold_result2 = cv.threshold(average2, 127, 255, cv.THRESH_BINARY)
threshold_value3, threshold_result3 = cv.threshold(average3, 127, 255, cv.THRESH_BINARY)
threshold_value4, threshold_result4 = cv.threshold(average4, 127, 255, cv.THRESH_BINARY)
threshold_value5, threshold_result5 = cv.threshold(average5, 127, 255, cv.THRESH_BINARY)

img_concate_Hori1=np.concatenate((threshold_result1,threshold_result2,threshold_result3),axis=1)
cv2_imshow(img_concate_Hori1)
img_concate_Hori2=np.concatenate((threshold_result4,threshold_result5),axis=1)
cv2_imshow(img_concate_Hori2)
```

## Adaptive Threshold

```
# Adaptive mean
threshold_result_1 = cv.adaptiveThreshold(average1, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
threshold_result_2 = cv.adaptiveThreshold(average2, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
threshold_result_3 = cv.adaptiveThreshold(average3, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
threshold_result_4 = cv.adaptiveThreshold(average4, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
threshold_result_5 = cv.adaptiveThreshold(average5, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)

img_concate_Hori1=np.concatenate((threshold_result_1,threshold_result_2,threshold_result_3),axis=1)
cv2_imshow(img_concate_Hori1)
img_concate_Hori2=np.concatenate((threshold_result_4,threshold_result_5),axis=1)
cv2_imshow(img_concate_Hori2)

# Adaptive Gaussian

thresholding_result_a = cv.adaptiveThreshold(average1, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
thresholding_result_b = cv.adaptiveThreshold(average2, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
thresholding_result_c = cv.adaptiveThreshold(average3, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
thresholding_result_d = cv.adaptiveThreshold(average4, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
thresholding_result_e = cv.adaptiveThreshold(average5, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)

img_concate_Hori1=np.concatenate((thresholding_result_a,thresholding_result_b,thresholding_result_c),axis=1)
cv2_imshow(img_concate_Hori1)
img_concate_Hori2=np.concatenate((thresholding_result_d,thresholding_result_e),axis=1)
cv2_imshow(img_concate_Hori2)
```
